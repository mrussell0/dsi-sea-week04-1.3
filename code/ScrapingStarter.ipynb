### Use beautifulsoup to pull the pokemon's name, number, and fighting type out of the elements in the table

```python
import urllib2
from bs4 import BeautifulSoup
import json
import pandas as pd
soup = BeautifulSoup(urllib2.urlopen('http://www.serebii.net/diamondpearl/pokemon.shtml').read())
```

### Put the pokemon data into a dataframe and print it

```python
print soup.body
```

```python
tables = soup.find_all('table', width="98%")
raw = {
    'names':[],
    'number':[],
    'type1':[],
    'type2':[]
}
for idx1, table in enumerate(tables):
    for idx2, rows in enumerate(table.find_all('tr')):
        if idx2 > 0:
            for idx3, columns in enumerate(rows.find_all('td')):
                if idx3 == len(rows.find_all('td')) -4:
                    raw['names'] +=[columns.text.split('\n')[1]]
                if idx3 == 0:
                    raw['number'] += [columns.text.replace('\r','').replace('\n','').replace('\t','').replace('#','')]
                if idx3 == len(rows.find_all('td')) -2:
                    key = columns.find('a').contents
                    raw['type1'] += [key[0]['src'].replace('/pokedex-dp/type/','').replace('.gif','')]  
                if idx3 == len(rows.find_all('td')) -1:
                    key = columns.find('a').contents
                    raw['type2'] += [key[0]['src'].replace('/pokedex-dp/type/','').replace('.gif','')]
                    
df = pd.DataFrame(raw)

df['fighting type'] = df['type1'] + '-' + df['type2']
df['fighting type'] = [x.replace('-na','') for x in df['fighting type']]

#df = df.drop(['type1','type2'], 1)
df.drop(['type1','type2'], 1, inplace=True)
```
